**5. Try to change parameters of decision tree and analyze how they affect the quality of model**

```python
# criterion

criterion = ['gini', 'entropy', 'log_loss']
result = []

for i in criterion:
    clf_test = tree.DecisionTreeClassifier(criterion=i, random_state=42)
    clf.fit(X_train, Y_train)

    result.append(clf.score(X_test, Y_test))

plt.bar(criterion, result, color='green', alpha=0.75)
plt.show()
print(max(result))
```

The function to measure the quality of a split. Supported criteria are “gini” for the Gini impurity and “log_loss” and “entropy” both for the Shannon information gain. There are no obvious differences in avarage performance.

```python
# splitter

splitter = ['best', 'random']
result = []

for i in splitter:
    clf_test = tree.DecisionTreeClassifier(splitter=i, random_state=42)
    clf.fit(X_train, Y_train)

    result.append(clf.score(X_test, Y_test))

plt.bar(splitter, result, color='green', alpha=0.75)
plt.show()
print(max(result))
```

The strategy used to choose the split at each node. Supported strategies are “best” to choose the best split and “random” to choose the best random split. Best is more easy to overfit in training set if there are too many samples. Ramdom has better ablity of generalization, but will perform worse if the data is too simple.

```python
# max_depth

max_depth = [i for i in range(5, 10)]
max_depth.append('None')
result = []

for i in max_depth:
    clf_test = tree.DecisionTreeClassifier(max_depth=i, random_state=42)
    clf.fit(X_train, Y_train)

    result.append(clf.score(X_test, Y_test))

max_depth = [str(i) for i in max_depth]
plt.bar(max_depth, result, color='green', alpha=0.75)
plt.show()
print(max(result))
```

The maximum depth of the tree. If None, then nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples. Deep tree may overfit and shallow tree may underfit. So, it's important to choose appropriate depth.

```python
# min_samples_split

min_samples_split = [i for i in range(2, 15)]
result = []

for i in min_samples_split:
    clf_test = tree.DecisionTreeClassifier(min_samples_split=i, random_state=42)
    clf.fit(X_train, Y_train)

    result.append(clf.score(X_test, Y_test))

plt.bar(min_samples_split, result, color='green', alpha=0.75)
plt.show()
print(max(result))
```

The minimum number of samples required to split an internal node. The purpose is to prune the branches to generalize model. Lead to the problem of underfit if the number is too high.
